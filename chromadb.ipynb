```python
from langchain_community.document_loaders import PyPDFLoader

# Load PDF file
loader = PyPDFLoader("Transformers for Natural Language Processing.pdf")

# Extract pages as documents
documents = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

docs = text_splitter.split_documents(documents)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",
    model_kwargs={"device": "cuda"},
    encode_kwargs={"normalize_embeddings": True}
)
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="chroma_store"
)

print("Chroma Vector DB created!")
```

```python
db = Chroma(
    persist_directory="chroma_store",
    embedding_function=embeddings
)
```

```python
query = "What is Natural langugage processing?"
results = db.similarity_search(query, k=3)

print(results[0].page_content)
```

```python
results = db.similarity_search_with_score(query)

for doc, score in results:
    print(score, doc.page_content)
```

```python
from langchain_community.document_loaders import Docx2txtLoader

loader = Docx2txtLoader("Sonu_patel_Assignment_Rephrased.docx")
documents = loader.load()

print(documents[0].page_content[:500])
```

```python
print(docs)
print(len(docs))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from tqdm import tqdm

splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100
)

docs = list(tqdm(splitter.split_documents(documents), desc="Splitting DOCX"))
```

```python
len(docs)
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=3)

for i, doc in enumerate(docs):
    print(f"\n------ Chunk {i+1} ------")
    print(doc.page_content)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",
    model_kwargs={"device": "cpu"},
    encode_kwargs={"normalize_embeddings": True}
)
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma(
    persist_directory="chroma_docx_store",
    embedding_function=embeddings
)

for doc in tqdm(docs, desc="Adding DOCX chunks to Chroma"):
    db.add_texts([doc.page_content], metadatas=[doc.metadata])

db.persist()
print("DOCX stored successfully!")
```

```python
db = Chroma(
    persist_directory="chroma_docx_store",
    embedding_function=embeddings
)
```

```python
query = "What is the objective of the AI medical chatbot?"
results = db.similarity_search(query, k=5)

print(results[2].page_content)
```

```python
from langchain_community.document_loaders import UnstructuredImageLoader

loader = UnstructuredImageLoader("DL.jpg")
documents = loader.load()

print(documents[0].page_content[:300])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from tqdm import tqdm

splitter = RecursiveCharacterTextSplitter(chunk_size=10, chunk_overlap=5)

docs = list(tqdm(splitter.split_documents(documents), desc="Splitting Image Text"))
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",
    model_kwargs={"device": "cpu"},
    encode_kwargs={"normalize_embeddings": True}
)
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma(
    persist_directory="chroma_image_store",
    embedding_function=embeddings
)

for doc in tqdm(docs, desc="Adding image chunks to Chroma"):
    db.add_texts([doc.page_content], metadatas=[doc.metadata])

db.persist()

print("Image OCR + Chroma vector database created!")
```

```python
query = "loss function in DL?"
results = db.similarity_search(query, k=2)

print(results[0].page_content)
```
