```python
# from langchain import LLMChain
# Create a chain for the title of our story
from langchain_core.runnables import RunnableLambda
template = """<s><|user|>
Create a title for a story about {summary}. Only return the title.<|end|>
<|assistant|>"""
title_prompt = PromptTemplate(template=template, input_variables=["summary"])

chain = title_prompt | llm | RunnableLambda(lambda x: {"title": x.content})
#title = LLMChain(llm=llm, prompt=title_prompt, output_key="title")
```

```python
result = chain.invoke({"summary": "a robot learning emotions"})
print(result["title"])
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain

# 1. Create the Groq LLM
llm = ChatGroq(
    model_name="llama-3.1-8b-instant",
    api_key="paste own"
)

# 2. Classic Prompt Template (old LangChain style)
prompt = PromptTemplate(
    input_variables=["input_prompt"],
    template="{input_prompt}"
)

# 3. Classic LLMChain
llm_chain = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True
)

# 4. Test it
output = llm_chain.invoke({"input_prompt": "Hi! My name is Sonu Patel. What is 1 + 1?"})
print(output)
```
